## **Which countries had won the cup ?**

```python
gold = world_cup["Winner"]
silver = world_cup["Runners-Up"]
bronze = world_cup["Third"]

gold_count = gold.value_counts()
silver_count = silver.value_counts()
bronze_count = bronze.value_counts()

podium_count = pd.concat([gold_count, silver_count, bronze_count], axis=1, sort=True).fillna(0)
podium_count.columns = ['WINNER', 'SECOND', 'THIRD']
podium_count['TOTAL'] = podium_count.sum(axis=1)
podium_count = podium_count.sort_values(by='TOTAL', ascending=False).drop('TOTAL', axis=1)

colors = ['#FFD700', '#C0C0C0', '#CD7F32']

ax = podium_count.plot(kind="bar", color=colors, figsize=(15, 6), fontsize=14, width=0.8, align='center', legend=True)
ax.legend(title='Podium Finish', labels=['Winner', 'Runner-Up', 'Third'], title_fontsize=14, fontsize=12)
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

plt.xlabel('Countries', fontsize=14)
plt.ylabel('Number of Podium Finishes', fontsize=14)
plt.title('Number of Podium Finishes by Country', fontsize=16)

plt.show()
```

## **Number of goal per country**

```python
home = matches[['Home Team Name', 'Home Team Goals']].dropna()
away = matches[['Away Team Name', 'Away Team Goals']].dropna()

goals_per_country = pd.concat([home.rename(columns={'Home Team Name': 'countries', 'Home Team Goals': 'goals'}),
                               away.rename(columns={'Away Team Name': 'countries', 'Away Team Goals': 'goals'})])

goals_per_country['goals'] = goals_per_country['goals'].astype('int64')

top_10_goals = goals_per_country.groupby('countries')['goals'].sum().nlargest(10)

colors = sns.color_palette('viridis', n_colors=len(top_10_goals))

plt.figure(figsize=(12, 6))
ax = top_10_goals.plot(kind="bar", color=colors, fontsize=14)

plt.xlabel('Countries', fontsize=14)
plt.ylabel('Number of Goals', fontsize=14)
plt.title('Top 10 Countries with the Most Goals', fontsize=16)

for i, v in enumerate(top_10_goals):
    ax.text(i, v + 10, str(v), fontsize=12, ha='center', va='bottom')

plt.xticks(rotation=45)

plt.show()
```

## **Match outcome by home and away teams**

```python
def get_labels(matches):

    labels = np.empty(matches.shape[0], dtype=object)

    home_team_win = matches['Home Team Goals'] > matches['Away Team Goals']
    labels[home_team_win] = 'Home Team Win'

    away_team_win = matches['Home Team Goals'] < matches['Away Team Goals']
    labels[away_team_win] = 'Away Team Win'

    draw = matches['Home Team Goals'] == matches['Away Team Goals']
    labels[draw] = 'DRAW'

    return labels
```

```python
matches['outcome'] = get_labels(matches)
```

```python
matches.head()
```

```python
mt = matches['outcome'].value_counts()
mt
```

```python
colors = ['#66b3ff', '#99ff99', '#ffcc99']

plt.figure(figsize=(5, 5))
plt.pie(mt, labels=mt.index, autopct='%1.1f%%', colors=colors, startangle=140, shadow=True)

circle = plt.Circle((0, 0), 0.4, color='white', edgecolor='black', linewidth=0.5)
plt.gca().add_artist(circle)

plt.title('Match Outcomes by Home and Away Teams', fontsize=10)
plt.axis('equal')

plt.show()
```
