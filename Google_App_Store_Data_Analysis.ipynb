**Checking the Nan value of the dataframe**

```python
df.isnull().sum()
```

```python
def get_dataframe_info(df):
    temp = pd.DataFrame(index=df.columns)
    temp['data_type'] = df.dtypes
    temp['null_count'] = df.isnull().sum()
    temp['unique_count'] = df.nunique()
    return temp
```

```python
get_dataframe_info(df)
```

**Checking the rows of the column content rating having null**

```python
df[df['Content Rating'].isnull()]
```

```python
df.dropna(subset = ['Content Rating'], inplace=True)
```

**Replacing the missing values of the rating column with the mode value of that entire column.**

```python
modeValueRating = df['Rating'].mode()
```

```python
modeValueRating[0]
```

```python
df['Rating'].fillna(value=modeValueRating[0], inplace = True)
```

**Creating a graphical visualization to represent the proportion of free and paid apps available on the Play Store.**

```python
plt.figure(figsize=(10, 10))

colors = sns.color_palette("pastel")
labels = df['Type'].value_counts(sort=True).index
sizes = df['Type'].value_counts(sort=True)
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.legend(labels, loc="best", title="App Type", fontsize="large")
plt.title('Percentage of Free Vs Paid Apps in the Store', size=20)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(12, 10))

type_cat_counts = df.groupby(['Category', 'Type']).size().unstack(fill_value=0)
sns.set_style('whitegrid')
type_cat_counts.plot(kind='barh', stacked=True, figsize=(12, 10), width=0.7, colormap='viridis')

plt.xlabel('Count', fontsize=14)
plt.ylabel('Category', fontsize=14)
plt.title('Distribution of App Types within Categories', size=20)
plt.legend(title='App Type', fontsize='large')
plt.tight_layout()
plt.show()
```

**Top app category with the highest number of installations**

Apps that have made the highest earning

```python
Paid_Apps_df = df[df['Type'] == 'Paid']
```

```python
earning_df = Paid_Apps_df[['App', 'Installs', 'Price']]
earning_df['Earnings'] = earning_df['Installs'] * earning_df['Price'];
```

```python
top_earning_apps = earning_df.sort_values(by=['Earnings', 'Price'], ascending=False).head(50)
```

```python
plt.figure(figsize=(15, 9))

sns.set_style('whitegrid')
colors = sns.color_palette("viridis", len(top_earning_apps))
sns.barplot(x='App', y='Earnings', data=top_earning_apps, palette=colors)

plt.xlabel("Apps", fontsize=14)
plt.ylabel("Earnings", fontsize=14)
plt.xticks(rotation=90)
plt.title("Top Earning Apps", fontsize=20)
plt.tight_layout()
plt.show()
```
